```python
import pandas as pd
from json import load
```

# Load the Data

```python
with open("database.json", "r") as json_file:
    dk_dict = load(json_file)
```

## We find that the data consists of three main tables:
- games  
- players 
- events  

```python
dk_dict.keys()
```

```python
games_df = pd.DataFrame(dk_dict['games'])
players_df = pd.DataFrame(dk_dict['players'])
events_df = pd.DataFrame(dk_dict['events'])
```

### Games

```python
games_df['date'] = pd.to_datetime(games_df.date)
games_df.head()
```

```python
games_df.iloc[3].tgURL
```

### Players

```python
players_df
```

### Events

```python
events_df['startDate'] = pd.to_datetime(events_df.startDate)
events_df['endDate'] = pd.to_datetime(events_df.endDate)
events_df.head()
```

# Exploration

## What do players use as their initials in the score screens?

```python
player_initials = pd.DataFrame(players_df.loc['initials'].dropna()[1:])
player_initials
```

## What is each player's high score?

```python
player_high_score_games = pd.DataFrame(games_df.groupby(by="player")[['score', '$id']].max().sort_values(by='score',ascending=False))
# player_high_scores = pd.DataFrame(games_df.groupby(by="player").score.max().sort_values(ascending=False))

player_high_score_games
```
